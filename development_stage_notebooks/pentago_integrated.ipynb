```python
### Apologies for the janky version control. This cell is the 'unified' pentago class per last night's meeting. Once we're happy here, we can delete the cells below this.

class pentago:
    """
    
    """

    def __init__(self, state = None):
        """Initializes the class reservation"""
        print('initializing')
        print(state)
        if state == None:
            self.state = state = np.zeros((6,6), dtype=np.int)
        self.history = []
        self.winner = None
        self.gameover = False
        self.player_turn = 1
    
    def current_board_state(self):
        #self.state = state
        #print("The current board state is: ", self.state)
        return copy.copy(self.state)

    def start_date(self):
        y = time
        self.game_time = y
        return game_time
    
    def game_history(self, player, move, cuad, rotatation):
        self.history.append((boardstate_to_ideal_key(self.state), player, move, cuad, rotatation))
        #return self.history
        
    def place(self, locate=None, marbel=None):
        """
        locate: is a tuple (x,y) with the position in the board
                x is the horizontal coordinate (0-5)
                y is the vertical coordinate (0-5)
        """
        self.locate = locate
        self.marbel = marbel
        
        if self.locate == None:
            print('No location Added')
        elif self.locate[0] > 5 or self.locate[0] < 0 or self.locate[1] > 5 or self.locate[1] < 0:
            print('Out of Board')
        else:
            x = self.locate[0]
            y = self.locate[1]
            if(self.state[x,y] == 0):
                self.state[x,y] = self.marbel
                return 1
            else:
                print('Error!! Position taken. Please select another position.')
                return None

                
    def rotate(self, cuad= None, orient=None):
        """
        Cuad: Select cuadrant.
              If 1, topleft
              if 2, top right
              if 3, bottom left
              if 4, bottom right
        orient: Select orientation
                if 1, clockwise
                if -1, counterclockwise
        """
        self.cuad = cuad
        self.orient = orient
        
        mr = np.zeros((3,3), dtype=np.int)
        if cuad == 1:
            c1 = (0,3)
            c2 = (0,3)
        elif cuad == 2:
            c1 = (0,3)
            c2 = (3,6)
        elif cuad == 3:
            c1 = (3,6)
            c2 = (0,3)
        elif cuad == 4:
            c1 = (3,6)
            c2 = (3,6)

        mr = self.state[c1[0]:c1[1],c2[0]:c2[1]]
        if orient == 1:
            mr = np.array(list(zip(*mr[::-1])))
        elif orient == -1:
            mr = np.array(list(zip(*mr[:,::-1])))
            
        self.state[c1[0]:c1[1],c2[0]:c2[1]] = mr      
        print('rotation completed.')  

    def find_winner(self, board_state):
        player1_win = False
        player_min1_win = False
        diagonal1 = board_state.diagonal()
        diagonal2 = np.fliplr(board_state).diagonal()
        winning_slices =  np.vstack([board_state[1:,:].T, board_state[:-1,:].T, # all columns
                              board_state[:,1:], board_state[:,:-1], # all rows
                              diagonal1[1:], diagonal1[:-1], # diagonal 1
                              diagonal2[1:],diagonal2[1:], # diagonal 2
                              board_state.diagonal(offset=1), board_state.diagonal(offset=-1), # diagonal offsets 
                              np.fliplr(board_state).diagonal(offset=1), np.fliplr(board_state).diagonal(offset=-1)] ) # diagonal offsets
        sums = np.dot(winning_slices, np.array([1,1,1,1,1]))
        if 5 in sums: player1_win = True
        if -5 in sums: player_min1_win = True
        if player1_win == True or player_min1_win == True:
            print("Player 1 winner?", player1_win, "Player -1 marble winner?", player_min1_win)
            self.gameover = True
            if player1_win == True:
                self.winner = 1
            elif player_min1_win ==True:
                self.winner = 0
            self.history.append(self.winner)
        return "Win"

    def check_gameover(self):
        if not 0 in self.state:
              self.gameover = True
              print("The game board is full!")
        
    def full_move(self, move, cuad, direction, player, dtype=np.int):
        if player != self.player_turn:
            print( "error, wrong player turn. No move taken.")
            return 'Error, wrong player turn.'
        if self.place(move, player) == None:
          return 'Error, place already taken'# maybe returns here if illegal move <-- Part of place method
        self.rotate(cuad, direction)
        self.game_history(move, player, cuad, direction)
        self.find_winner(self.state) #return in find_winner if a winner is found
        self.check_gameover() #return in check_gameover
        if player == 1:
            self.player_turn = -1
        else:
            self.player_turn = 1
        print('Successful Move')
        return self.state


```

```python
%autoreload 2
class qtable_agent:
    
    def __init__(self, player = 1, epsilon = 1, epsilon_decay = .995, q_table = q_table()):
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.q_table = q_table
        self.player = player
        
            
    def get_avail_moves(self,boardstate):
        """
        This method creates a list with available spaces in the board and combination of quadrant and rotation
        The input is the board state (6x6) numpy array
        """
        x = np.where(boardstate == 0)
        #print(x)
        available_positions_for_placement = list(zip(x[0], x[1]))
        
        # all available positions (p), quadrants(q), rotations(r)
        available_moves = [(p,q,r) for p in available_positions_for_placement for q in [1,2,3,4] for r in [-1,1]]
        #print(len(available_moves))
        return available_moves
    
    def get_possible_next_boardstates(self, boardstate):
        next_possible_boardstates = defaultdict(list)
        for move in self.get_avail_moves(boardstate):
            possible_boardstate = fullmove(boardstate,*move, self.player)
            key = boardstate_to_ideal_key(possible_boardstate)
            #print(key)
            next_possible_boardstates[key].append(move)
            
        return next_possible_boardstates
    
    def make_move(self, game):
        
        # get the current boardstate from the pentago class
        boardstate = game.current_board_state()
        
        # get possible next possible boardstates
        next_possible_boardstates = self.get_possible_next_boardstates(boardstate)
        key_list = list(next_possible_boardstates.keys())
        
        # determine if to take random move
        if np.random.rand() < self.epsilon:
            random_bs = random.choice(key_list)
            random_mv = next_possible_boardstates[random_bs][0]
            
            game.full_move(*random_mv,self.player)
            
        else:
            print("not random")
            q_values_list = [self.q_table.get_q_value(bs) for bs in key_list]
            print(q_values_list)
            index_of_best_move = q_values_list.index(max(q_values_list))
            
            ## list all max q_values and choose at random
            mv_to_take = next_possible_boardstates[key_list[index_of_best_move]][0]
            game.full_move(*mv_to_take, self.player)
            
        self.epsilon *= self.epsilon_decay 
```

```python
## initial code from Ernesto

class agent:
    
    def __init__(self):
        self.action = None
        self.av = []
        self.new_state = np.zeros((6,6), dtype=np.int)
      
    def avail(self,state):
        """
        This method creates a list with available spaces in the board
        The input is the board state (6x6) numpy array
        """
        for i in range(6):
            for j in range(6):
                if state[i,j]==0:
                    self.av.append((i,j))
                    
    def play(self, state,q_indic, eps = 1.0, mbl = 1):
        """
        play method place a marbel by an imput state of the board
        q_indic is the dictionary with q values in the followinf format:
        {board code (string): q value (float)}, For example:
        {'212201122201222112201022110121120022': 0.65}
        eps: is epasilon. It is the threshold to ramdomly determine to
             make a random move or play by q values
        mbl: 1 or 2. the marbel playing.
        
        """
        
        def mat2str(m):
            """Board matrix to string code"""
            v = m.flatten()
            st = ''
            for k in v:
                st += str(k)
            return st
        
        def str2mat(code):
            """String code to matrix"""
            v = [int(k) for k in code]
            npv = np.array(v)
            return npv.reshape(6,6)
        
        #Start a petago object to analize the moves
        test_table = pentago(np.zeros((6,6), dtype=np.int))
        
        if eps > random.uniform(0,1): # Play random
            xy = self.av[random.randint(0,len(self.av)-1)] #Select from availables spots 
            c = random.randint(1,4)
            r = random.choice([-1, 1])                                     
            self.action = (xy, mbl, c, r)
            test_table.state = state.copy() # Copy the board state
            test_table.full_move(xy,mbl,c, r) # Full Move
            self.new_state = test_table.state
            
        else:# Play acording to q values
            act_dic = {}
            q_dic = {}
            
            #Creating the states after all possible move and creaste 2 dictionaries
            #with resultant states codes, actions and qvalues 
            for k in self.av:
                for j in range(1,5):
                    for i in [-1,1]: 
                        test_table.state = state.copy()                        
                        test_table.full_move(k,mbl,j,i)
                        state_code = mat2str(test_table.state)
                        try:
                            q_dic[state_code] = q_indic[state_code]     
                            act_dic[state_code] = (k,mbl,j,i)
                        except:
                            pass
            
            #If q values are found, select the maximum value
            #If more than one qvalue is maximum, pick one randomly
            if len(act_dic) > 0:
                max_value = max(q_dic.values())
                [k for k,v in q_dic.items() if v == max_value]
                code = random.choice(k)
                self.action = act_dic[code]
                self.new_state = str2mat(code)
                
            #If no qvalue is found, play randomly
            else:
                xy = self.av[random.randint(0,len(self.av)-1)]
                c = random.randint(1,4)
                r = random.choice([-1, 1])                                     
                self.action = (xy, mbl, c, r)
                test_table.state = state.copy()
                test_table.full_move(xy,mbl,c, r)
                self.new_state = test_table.state
```

```python
class play_pentago(object):
    def run(self):
        
        table = pentago(np.zeros((6,6), dtype=np.int))
        mb=1
        cuadrants = [1,2,3,4]
        rotation = [-1,1]
        av = [(i,j) for i in range(6) for j in range(6)]
        
        cont = 1
        while av != []:
            
            random.shuffle(av) 
            xy = av.pop()
            
            table.place(xy,mb)            
            print(table.state,'Marbel Placed','\n')

            random.shuffle(cuadrants) 
            random.shuffle(rotation)
            table.rotate(cuadrants[0],rotation[0])
            print(table.state,'Rotation','\n')
            print('Play: {}, Marbel: {}, Placed: {}, Cuadrant: {}, Rotated: {}\n'
                  .format(cont,mb,xy,cuadrants[0],rotation[0]))
            
            mb = -mb # Switching player
            
            av = [] # Available list
            for i in range(6):
                for j in range(6):
                    if table.state[i,j]==0:
                        av.append((i,j))
                        
            cont += 1
        return #Main program
```
